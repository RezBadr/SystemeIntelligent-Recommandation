```python

import sqlite3
from sqlalchemy import create_engine  

conn = sqlite3.connect('book_reviews.db')  
# list tables
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print(tables)
df = pd.read_sql_query("SELECT * FROM book_reviews LIMIT 1000;", conn)
for col in ['likes_on_review','reviewer_followers','reviewer_total_reviews','review_rating']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
print(df.head())
conn.close()

engine = create_engine('sqlite:///books.db')  
with engine.connect() as conn:
    tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
    print(tables)
    df2 = pd.read_sql_query("SELECT * FROM books LIMIT 1000;", conn)
    for col in ['book_ids','title','total_books_in_collection','total_votes_for_collection']:
        if col in df2.columns:
            df2[col] = pd.to_numeric(df2[col], errors='coerce')
    print(df2.head())
```
